## Task D2: LOOCV

Define the function `calculate_loocv(data, model, dependent_var)` so that the `task4_c` function works correctly.
It should use one observation from `data` for testing and the rest of the data for training. For each model calculate the mean squared error and return the mean error across all the errors for all models.

```python
# implements Rs poly function
def poly(x, p):
    x = np.array(x)
    X = np.transpose(np.vstack([x ** k for k in range(p + 1)]))
    X = np.linalg.qr(X)[0][:, 1:]
    return X
```

```python
from sklearn.model_selection import LeaveOneOut, cross_val_score
```

```python
def calculate_loocv(data, model, dependent_var):
    leave_validation=LeaveOneOut()
    result = cross_val_score(model,data,data["y"],cv=leave_validation)
    return result
```

```python
def task4_c(seed=1):
    # 8 c)
    data = task4_a()
    np.random.seed(seed)
    print("Y=β0+β1X+ε")
    print(calculate_loocv(data, LinearRegression(), "y"))

    temp = pd.DataFrame(poly(data["X"], 2), columns=["X", "X_2"])
    temp["y"] = data["y"]
    print("Y=β0+β1X+β2X2+ε")
    print(calculate_loocv(temp, LinearRegression(), "y"))

    temp = pd.DataFrame(poly(data["X"], 3), columns=["X", "X_2", "X_3"])
    temp["y"] = data["y"]
    print("Y=β0+β1X+β2X2+β3X3+ε")
    print(calculate_loocv(temp, LinearRegression(), "y"))

    temp = pd.DataFrame(poly(data["X"], 4), columns=["X", "X_2", "X_3", "X_4"])
    temp["y"] = data["y"]
    print("Y=β0+β1X+β2X2+β3X3+β4X4+ε")
    print(calculate_loocv(temp, LinearRegression(), "y"))

task4_c()
```
