```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import statsmodels.formula.api as smf
```

```python
def task7_data(seed=42):
    np.random.seed(seed)
    df = pd.DataFrame()
    df["X1"] = np.random.uniform(low=0.0, high=1.0, size=100)
    df["X2"] = 0.5 * df["X1"] + np.random.normal(0, 1, 100) / 10
    df["y"] = 2 + 2 * df["X1"] + 0.3 * df["X2"] + np.random.normal(0, 1, 100)
    return df

data = task7_data()
data
```

Plot the correlation between $X_1$ and $X_2$ as a scatterplot.
