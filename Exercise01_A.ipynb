## Task 5: Creating _Elites_

Create a new column `Elite` by _binning_ the `Top10perc` column.
Divide the universities into two groups based on wheter or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.

Then, similar to above, produce a side-by-side boxplot with the newly created indicator `Elite` as the _x_-axis and the `Outstate` as _y_.

After the plot, print the number of universities which are "elite" based on the criteria above.

```python
# YOUR CODE HERE
df["Elite"] = df["Top10perc"].apply(lambda k: k > 50)
df["Elite"]

sns.boxplot(
    x="Elite",
    y="Outstate",
    data = df
)
```
