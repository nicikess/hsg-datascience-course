## Task 1: Decision Boundary

In the lecture you have seen this formula for the discriminant score:

$$
\delta_k(x) = x * \frac{\mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + log(\pi_k)
$$

Show that for $K=2$ classes and $\pi_1 = \pi_2 = 0.5$ the decision boundary is at (on paper, not programmatically):

$$
x = \frac{\mu_1+\mu_2}{2}
$$
