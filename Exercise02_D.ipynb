In this exercise we use randomly generated data. We will investigate the t-statistic for the null hypothesis $H_0 : \beta = 0$ in simple linear regresion without an intercept.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def task4_data(seed=42):
    np.random.seed(seed)
    X = np.random.normal(0, 1, 100)
    y = 2 * X + np.random.normal(0, 1, 100)
    return X, y

X, y = task4_data()
```

## Task D1: Linear Regression: y onto x

Fit a linear regression of y onto x, without intercept, and analyze the results.

```python
# dont add constant to get lr. without an intercept
fitted = sm.OLS(X,y).fit()
fitted.summary()
```
